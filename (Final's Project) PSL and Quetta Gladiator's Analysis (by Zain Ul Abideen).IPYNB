The analysis includes

- Exploratory Data Analysis (EDA)
- Data Visualization

### Importing Libraries and Dataset required for Analysis
For loading the dataset and to apply Analytical techniques, I used the two libraries: *Pandas* and *Numpy*

```python
import numpy as np
import pandas as pd

psl = pd.read_csv("Matches Detail PSL 2016- 2024.csv")


psl
```

### Data Manipulation and Cleaning
In this step, I
- renamed some wrongly spelled column names, and also changed a column name according to my need
- dropped some unwanted columns, and a repeating column
- renamed the values (string) inside a column, removing the issues present in their name, corrected them to make accurate unique values

```python
psl.rename(columns={
    "Match Numer": "Match Number",
    "Win by Wckets": "Win by Wickets",
    "Ground": "Venue"
}, inplace=True)

psl.drop(columns=['Match days', 'Umpires', 'TV Umpire', 'Reserve Umpire', 'Match Referee'], inplace=True)

psl['Match Date'] = pd.to_datetime(psl['Match Date'], format='%d-%b-%y')

psl['Day & Night'] = psl['Day & Night'].str.title()
psl['Day & Night'] = psl['Day & Night'].replace({
    'Daynight': 'Day & Night',
    ' Daynight ': 'Day & Night'
})

psl
```

### Calculatimg the Toss Impact on all Matches

```python
toss_impact_list = []

for i in range(len(psl)):
    if psl.loc[i, 'Toss Winner'] == psl.loc[i, 'Winner']:
        toss_impact_list.append('Toss Winner Won')  
    else:
        toss_impact_list.append('Toss Winner Lost') 

psl['Toss Impact'] = toss_impact_list

psl
```

## Filtering Quetta Gladiators Matches
As a huge fan of Quetta Gladiators myself, I decided to do *EDA* on my faviorite club

In the very first step, I filtered out Quetta Gladiators matches and created Yearly Summary 

The yearly summary includes the total_matches, wins, losses, no_results, valid_matches, and win_percentage of Gladiators throughout every season of PSL

```python
quetta_matches = psl[(psl['Team 1'] == 'Quetta Gladiators') | (psl['Team 2'] == 'Quetta Gladiators')]

# quetta_matches['Year'] = pd.to_datetime(quetta_matches['Match Date']).dt.year
quetta_matches
yearly_summary = []
for year in (quetta_matches['Season'].unique()):
    year_data = quetta_matches[quetta_matches['Season'] == year]
    total_matches = len(year_data)
    wins = len(year_data[year_data['Winner'] == 'Quetta Gladiators'])
    no_results = len(year_data[year_data['Result'] == 'No Result'])
    valid_matches = total_matches - no_results
    losses = valid_matches - wins
    win_percentage = round((wins / valid_matches) * 100, 2)
    
    yearly_summary.append([year, total_matches, wins, losses, no_results, valid_matches, win_percentage])

```

### Displaying the yearly summary as a DataFrame

```python
yearly_summary_df = pd.DataFrame(yearly_summary, columns=['Season', 'Total Matches', 'Wins', 'Losses', 'No Results', 'Valid Matches', 'Win Percentage'])
yearly_summary_df
```

### Visualizing theyearly summary by Plotting **Win Percentage Over Seasons**

For visualization (plotting graphs), I used the libraries, *Matplotlib* and *Seaborn*

I plotted Quetta Gladiator's win percentage over the years using a **line chart** for **trend visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.lineplot(
    data = yearly_summary_df,
    x = 'Season',
    y = 'Win Percentage',
    marker = 'o',
    color = 'red'
)

plt.title('Quetta Gladiators Win Percentage Over Seasons')
plt.xlabel('Season')
plt.ylabel('Win Percentage')
plt.grid()
plt.tight_layout()

plt.show()
```

### Toss Impact Analysis Per Season for Quetta Gladiators

In this cell, I analyzed how often Quetta Gladiators won the match after winning the toss for each season.

```python
toss_impact_data = []
for year in (quetta_matches['Season'].unique()):
    year_data = quetta_matches[quetta_matches['Season'] == year]
    toss_wins = len(year_data[year_data['Toss Winner'] == 'Quetta Gladiators'])
    toss_win_and_match_win = len(year_data[(year_data['Toss Winner'] == 'Quetta Gladiators') & (year_data['Winner'] == 'Quetta Gladiators')])
    toss_impact = round((toss_win_and_match_win / toss_wins) * 100, 2)
    toss_impact_data.append([year, toss_wins, toss_win_and_match_win, toss_impact])

toss_impact_df = pd.DataFrame(toss_impact_data, columns=['Season', 'Toss Wins', 'Toss Win & Match Win', 'Toss Impact %'])
toss_impact_df
```

### Visualizing Toss Impact Over Seasons

For this, I plotted toss impact percentage per season, showing how beneficial toss wins have been for Quetta Gladiators.

```python
plt.figure(figsize=(10,6))
sns.lineplot(
    data = toss_impact_df,
    x = 'Season',
    y = 'Toss Impact %',
    marker = 'o',
    color = 'blue'
)

plt.title('Quetta Gladiators Toss Impact Over Seasons')
plt.xlabel('Season'),
plt.ylabel('Toss Impact %'),
plt.grid()

plt.show()

```

### Venue-Based Stats Collection

This breaks down Quetta’s performance venue-wise, recording total matches, wins, and losses at each ground.

```python
venues = quetta_matches['Venue'].unique()
venue_names = []
matches_played = []
matches_won = []
matches_lost = []
win_percentages = []

for venue in venues:
    venue_data = quetta_matches[quetta_matches['Venue'] == venue]
    total = len(venue_data)
    wins = len(venue_data[venue_data['Winner'] == 'Quetta Gladiators'])
    losses = len(venue_data[(venue_data['Winner'] != 'Quetta Gladiators') & (venue_data['Result'] != 'No Result')])
    win_percentage = round((wins / total) * 100, 2) 

    venue_names.append(venue)
    matches_played.append(total)
    matches_won.append(wins)
    matches_lost.append(losses)
    win_percentages.append(win_percentage)

venue_stats_quetta = pd.DataFrame({'Venue': venue_names, 'Total Matches': matches_played, 'Wins': matches_won, 'Losses': matches_lost, 'Win Percentage': win_percentages})
venue_stats_quetta


```

### Venue-wise Wins Bar Chart

To show how many matches Quetta Gladiators have won at each venue, I plotted a Bar plot

```python
plt.figure(figsize=(10, 6))
sns.catplot(
    data = venue_stats_quetta,
    x = 'Venue',
    y = 'Wins',
    palette = 'viridis',
    kind='bar',
    height=6,
    aspect=1.5
)

plt.title('Quetta Gladiators Wins by Venue')
plt.xlabel('Venue')
plt.ylabel('Wins')
plt.xticks(rotation=45)

plt.show()
```

### Head-to-Head Analysis vs Other Teams

I made a detailed comparison of Quetta Gladiators’ head-to-head record against each club including wins, losses, and win percentages.

```python
quetta_matches = psl[(psl['Team 1'] == 'Quetta Gladiators') | (psl['Team 2'] == 'Quetta Gladiators')]


head_to_head_matches = []
head_to_head_wins =[]
head_to_head_losses = []
head_to_head_no_results = []
head_to_head_win_percentage = []
total_no_of_matches = []

for opponent in sorted(quetta_matches['Team 1'].unique()):
    if opponent == 'Quetta Gladiators':
        continue  
    else:
        opponent_matches = quetta_matches[
            (quetta_matches['Team 1'] == opponent) | (quetta_matches['Team 2'] == opponent)
        ]
    
    total_matches = len(opponent_matches)
    wins = len(opponent_matches[opponent_matches['Winner'] == 'Quetta Gladiators'])
    no_results = len(opponent_matches[opponent_matches['Result'] == 'No Result'])
    losses = total_matches - wins - no_results
    win_percentage = round((wins / (total_matches - no_results)) * 100, 2) 
    total_no_of_matches_a = wins + losses + no_results

    head_to_head_matches.append(opponent)
    head_to_head_wins.append(wins)
    head_to_head_losses.append(losses)
    head_to_head_no_results.append(no_results)
    head_to_head_win_percentage.append(win_percentage)
    total_no_of_matches.append(total_no_of_matches_a)

 
head_to_head_matches
head_to_head_wins
head_to_head_losses
head_to_head_no_results
head_to_head_win_percentage


head_to_head_df = pd.DataFrame({
    'Opponent': head_to_head_matches,
    'Matches' : total_no_of_matches,
    'Wins' : head_to_head_wins,
    'Losses': head_to_head_losses,
    'No Results': head_to_head_no_results,
    'Win Percentage': head_to_head_win_percentage
})

head_to_head_df
```

### Head-to-Head Win Percentage Bar Plot

I plotted win percentage vs opponents. It helped me to visualize which teams Quetta Gladiators have performed well or struggled against.

```python
plt.figure(figsize=(10, 4))
sns.barplot(
    data=head_to_head_df,
    x='Win Percentage',
    y='Opponent',
    palette='viridis',
)

plt.title('Quetta Gladiators Head-to-Head Win Percentage')
plt.xlabel('Win Percentage (%)')
plt.ylabel('Opponent')
# plt.xlim(0, 100)
plt.xticks(range(0, 101, 10))
# plt.show()

plt
```
